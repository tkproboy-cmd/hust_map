```python
import geopandas as gpd
buildings = gpd.read_file('hust_building.geojson') #done
grass = gpd.read_file('grass.geojson') #done
water = gpd.read_file('water.geojson') #done
base = gpd.read_file('hust_base.geojson') #done
parkings = gpd.read_file('parking_clean.geojson') #done
roads = gpd.read_file('road_cleaner_detail.geojson') #done
trees = gpd.read_file('tree_point.geojson') # not done
import folium
m = folium.Map(location=[21.005, 105.843], zoom_start=17)
```

```python
import folium
#Bound
hust_bounds = [[21.000, 105.835], [21.010, 105.855]]
m = folium.Map(
    location=[21.005, 105.843],
    zoom_start=16,
    tiles=None,              
    min_zoom=16,             
    max_zoom=18,             
    max_bounds=True          
)
m.fit_bounds(hust_bounds)
m.options['maxBounds'] = hust_bounds
# thêm lớp base
folium.GeoJson(
    base,
    name="Nền trường Bách Khoa",
    style_function=lambda x: {
        'fillColor': "#F8F9FA", 
        'color': '#D3D3D3',
        'weight': 1,
        'fillOpacity': 0.3
    }
).add_to(m)
m.save('hust_map.html')
```

```python
# lớp roads
folium.GeoJson(
    roads,
    name="Đường đi",
    style_function=lambda x: {
        'fillColor': "#E9DAD5",
        'color': "#4C5455",
        'weight': 0.5,
        'fillOpacity': 0.7
    }
).add_to(m)
m.save('hust_map.html')
```
